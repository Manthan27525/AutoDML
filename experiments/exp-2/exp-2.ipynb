```python
import pandas as pd
from autodml import Autodml
```

```python
df=pd.read_csv("../../temp/bmw_sales.csv")
df.head()
```

```python
model=Autodml(data=df,target="Avg_Price_EUR")
model.train()
```

```python
model.predict({
      "Year": 2024,
      "Month": 6,
      "Region": "Europe",
      "Model": "BMW i4",
      "Units_Sold": 1200,
      "Revenue_EUR": 72000000.0,
      "BEV_Share": 0.35,
      "Premium_Share": 0.60,
      "GDP_Growth": 1.8,
      "Fuel_Price_Index": 1.25})
```
